```python
import numpy as np
import pandas as pd
```

```python
#Question 2: Create a Series and DataFrame using the data given in Table 2.14.

data1= {"Location":
 ['United States', 'India', 'Brazil', 'Russia', 'United Kingdom', 'France', 'Spain', 'Italy', 'Turkey', 'German'],
      "Cases":
       [ 28833039, 11079979, 10517232, 4246079, 4170519, 3736016, 3188553, 2907825, 2693164, 2444169],
    "Deaths":
      [517204, 156938, 254263, 86122, 122705,86332, 69142, 97507, 28503, 70601 ],
    "Recovered":
     [np.nan, 10763451, 9386440, 3811797, np.nan, np.nan, np.nan, 2398352, 2565723, 2242767]
      }


df1= pd.DataFrame(data1)

series_cases= pd.Series(df1["Cases"].values, index= df1["Location"], name= "Cases")
series_deaths= pd.Series(df1["Deaths"].values, index= df1["Location"], name= 'Deaths')
series_recovered= pd.Series(df1["Recovered"].values, index= df1["Location"], name= 'Recovered')

print(df1)
print(series_cases)
print(series_deaths)
print(series_recovered)
```

```python
#Question 3: Create a new column (Status) in Qs 2 Dataframe. Analyze the “Cases” column, in case of less than 3 million cases set the Status to low. In case of greater than 3 million and less than 4 million cases Status is medium, otherwise high.

def status_col(cases):
  if cases < 30000000:
    return 'Low'
  elif 30000000 <= cases > 40000000:
    return 'Medium'
  else:
    return 'High'

df1["Status"]= df1['Cases'].apply(status_col)
df1
```

```python
#Question 4:Rank the created Series and DataFrame in ascending order of deaths count.

df1new= df1.fillna(0)
cases_rank= series_cases.rank()
death_rank= series_deaths.rank()
recovered_rank= series_recovered.rank()

df1new['Deaths_Rank']= df1['Deaths'].rank(method= 'max')
df1new
```
